```python
grades = []
for x in df.columns[:-1]:
    for y in df[x].unique():
        if y not in grades:
            grades.append(y)
print(grades)
```

```python
#We need to replace the grades with some numerical values
```

```python
grades_enc = {'A+':4,
              'A':4,
              'A-':3.7,
              'B+':3.4,
              'B':3.0,
              'B-':2.7, 
              'C+':2.4, 
              'C':2.0, 
              'C-':1.7,
              'D+':1.4,
              'D':1.0, 
              'WU':0.0, 
              'F':0.0,
              'W':0,
              'I':0}
```

```python
#For every better grade we add 0.3 or 0.4 
```

```python
for x in df.columns[:-1]:
    df[x] = df[x].replace(grades_enc)
```

```python
X = df.drop('CGPA',axis=1)
y = df['CGPA']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr_model = LinearRegression()
```

```python
lr_model.fit(X_train,y_train)
```

```python
lr_pred = lr_model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
sns.scatterplot(x=lr_pred,y=y_test)
```

```python
from sklearn.metrics import r2_score
```

```python
lr_pred
```

```python
r2_score(y_test,lr_pred)
```

```python
#We have got a r2_score of 0.96
```

```python
mean_absolute_error(y_test,lr_pred)
```

```python
mean_squared_error(y_test,lr_pred)
```

```python
np.sqrt(mean_squared_error(y_test, lr_pred))
```
